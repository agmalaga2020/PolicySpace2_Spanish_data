### üìÇ Archivo: `average_num_members_families_2010.csv`

üîç **¬øA qu√© hace referencia exactamente?**  
Este archivo representa el tama√±o medio de los hogares, es decir, el n√∫mero promedio de miembros por familia que residen en viviendas particulares. Se trata de una de las variables base que utiliza *PolicySpace2* para simular din√°micas demogr√°ficas y de demanda de vivienda. Es una medida importante porque afecta a la formaci√≥n de nuevos hogares, la presi√≥n sobre el mercado inmobiliario y la distribuci√≥n del gasto familiar.

üåç **¬øC√≥mo est√°n definidos geogr√°ficamente?**  
En *PolicySpace2*, esta variable se define para cada una de las regiones metropolitanas simuladas en Brasil (46 en total). Los datos provienen del censo brasile√±o de 2010, y se utiliza el valor promedio de tama√±o de familia para iniciar la generaci√≥n de agentes (households) en cada regi√≥n.

üõ† **¬øC√≥mo adaptarlo al contexto Espa√±ol?**  
Nuestros datos del INE permiten replicar esta variable f√°cilmente. Ya que en Espa√±a contamos con estad√≠sticas del tama√±o medio de los hogares por **comunidad aut√≥noma**, podemos utilizarlas como proxy directa para inicializar los agentes en nuestro modelo. Dado que *PolicySpace2* trabaja con regiones metropolitanas, y nuestros datos est√°n agregados por comunidades, una opci√≥n es:

- Asignar directamente el valor de la comunidad aut√≥noma a todas las provincias dentro de ella.
- O bien, si se quieren afinar m√°s las simulaciones, desagregar por provincias o √°reas funcionales urbanas si se dispone de esos datos m√°s desglosados.

üß† **Conclusi√≥n para el proyecto**  
Podemos adaptar perfectamente esta variable al modelo espa√±ol sin perder riqueza descriptiva. Basta con usar el valor del tama√±o medio de hogar para cada comunidad aut√≥noma seg√∫n el INE (a√±o 2021 o m√°s reciente), como punto de partida para la simulaci√≥n de hogares. ### extraemos las tablas

```python
import requests
import pandas as pd

# C√≥digo de la tabla
table_code = "60132"
url_csv = f"https://servicios.ine.es/wstempus/csv/ES/DATOS_TABLA/{table_code}?nult=999"

response = requests.get(url_csv)

if response.status_code == 200:
    csv_data = response.content.decode('utf-8')

    # Guardar el archivo CSV con un nombre adecuado usando el c√≥digo de la tabla
    file_name = f"tabla_{table_code}.csv"
    with open(file_name, "w", encoding="utf-8") as f:
        f.write(csv_data)
    print(f"Archivo CSV guardado como '{file_name}'")
else:
    print("Error en la petici√≥n: ‚ùå", response.status_code)

# Leer el CSV
df = pd.read_csv(file_name, sep='\t', encoding='utf-8')

# Mostrar las primeras filas
print(df.head())
print("Tabla actualizada conseguida ‚úÖ")

try:
    df.to_csv(f'/content/{table_code}.csv', index=False)
    print(f"‚úÖ Copia de df guardada en '/content/{table_code}.csv'")
    print("Se ha guardado una copia del DataFrame df.")
except NameError:
    print("‚ö†Ô∏è df o table_code no est√°n definidos. No se puede guardar la copia.")
```

```python
# Usar la funci√≥n de exploraci√≥n definida para ver detalles de las columnas
def print_unique_values_and_len(df):
    """Imprime valores √∫nicos, n√∫mero de NaN y tipo por columna."""
    for column in df.columns:
        unique_values = df[column].unique()
        nan_count = df[column].isnull().sum()
        dtype = df[column].dtype
        print(f"üß© Columna: {column}")
        print(f"- Valores √∫nicos: {unique_values}")
        print(f"- Total √∫nicos: {len(unique_values)}")
        print(f"- NaNs: {nan_count}")
        print(f"- Tipo: {dtype}\n")

print_unique_values_and_len(df)
```

```python
import pandas as pd
import warnings

# Suprime los warnings
warnings.filterwarnings('ignore')

# URL y lectura de la primera tabla
url = 'https://www.ine.es/daco/daco42/codmun/cod_ccaa_provincia.htm'
df_temp = pd.read_html(url, encoding='ISO-8859-1')[0]
df_temp.columns = ['CODAUTO', 'Comunidad Aut√≥noma', 'CPRO', 'Provincia']

# Filtra para eliminar "Ciudades Aut√≥nomas" y convierte CODAUTO a num√©rico
df_temp = df_temp[df_temp['Comunidad Aut√≥noma'] != 'Ciudades Aut√≥nomas']
df_temp['CODAUTO'] = pd.to_numeric(df_temp['CODAUTO'], errors='coerce')
df_temp = df_temp.dropna(subset=['CODAUTO'])

# Guarda el DataFrame en CSV y muestra las primeras filas
df_temp.to_csv('codigos_ccaa_provincias.csv', index=False, encoding='utf-8')
print(df_temp.head())

# Crea y muestra el DataFrame de equivalencias de comunidades
equivalencias_comunidades = df_temp[['CODAUTO', 'Comunidad Aut√≥noma']].drop_duplicates('CODAUTO')
print(equivalencias_comunidades)
```


- ‚úÖ `Comunidades y ciudades aut√≥nomas` ‚Üí `comunidad_name`
- ‚úÖ `Periodo` separado en `a√±o` y `mes` (formato num√©rico de mes)
- ‚úÖ `Total` convertido a tipo `float` como `total`
- ‚úÖ A√±adida columna `comunidad_code` a partir de `equivalencias_comunidades.csv`



```python
# Copia del DataFrame original para limpieza
df_clean = df.copy()

# Renombrar columna
df_clean.rename(columns={'Comunidades y ciudades aut√≥nomas': 'comunidad_name'}, inplace=True)

# Extraer a√±o y mes de la columna 'Periodo'
meses_dict = {
    'enero': '01', 'febrero': '02', 'marzo': '03', 'abril': '04',
    'mayo': '05', 'junio': '06', 'julio': '07', 'agosto': '08',
    'septiembre': '09', 'octubre': '10', 'noviembre': '11', 'diciembre': '12'
}

# Crear nuevas columnas 'a√±o' y 'mes'
df_clean['mes'] = df_clean['Periodo'].str.extract(r'de (\w+) de')[0].str.lower()
df_clean['mes'] = df_clean['mes'].map(meses_dict)
df_clean['a√±o'] = df_clean['Periodo'].str.extract(r'de (\d{4})')[0].astype(int)

# Convertir la columna 'Total' a formato num√©rico (cambiando coma por punto)
df_clean['total'] = df_clean['Total'].str.replace(',', '.').astype(float)

# Unir con dataset de equivalencias para obtener comunidad_code
df_clean = df_clean.merge(equivalencias_comunidades, left_on='comunidad_name', right_on='Comunidad Aut√≥noma', how='left')
df_clean.rename(columns={'CODAUTO': 'comunidad_code'}, inplace=True)

# Eliminar columnas no necesarias
df_clean.drop(columns=['Periodo', 'Total', 'Comunidad Aut√≥noma'], inplace=True)

# Eliminar filas donde comunidad_name es 'Total Nacional'
df_clean = df_clean[df_clean['comunidad_name'] != 'Total Nacional']

# Mostrar dimensiones despu√©s del filtrado
v.shape

```
