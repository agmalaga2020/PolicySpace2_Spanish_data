## Fuentes de datos por municipio (2000‚Äì2024)

Para obtener el nivel de **cualificaci√≥n educativa de la poblaci√≥n por municipio** en Espa√±a, las fuentes m√°s confiables son los **censos de poblaci√≥n** y estad√≠sticas oficiales derivadas, complementadas por datos auton√≥micos y portales abiertos:

- **Censos de Poblaci√≥n (INE)** ‚Äì Los censos nacionales proporcionan puntos de datos clave (aprox. decenales) con **detalle municipal**. En el **Censo 2001** y **Censo 2011** se public√≥ la poblaci√≥n por nivel de estudios alcanzado en cada municipio (habitualmente para poblaci√≥n de 16 o 15 a√±os y m√°s, desglosada por categor√≠as educativas) ([Censos de Poblaci√≥n y Viviendas 2001 - INE](https://www.ine.es/censo/es/listatablas.jsp?group=5.Tablas%20comparativas%20de%20municipios%3E1.Personas%3E3.Estudios%20y%20estudiantes%3E0.Todos%20los%20municipios#:~:text=0.Todos%20los%20municipios.%20,la%20tabla%20en%20formato%20Excel)). El INE difundi√≥ tablas descargables por municipio; p. ej., para 2001 existe una tabla ‚ÄúTodos los municipios ‚Äì poblaci√≥n de 16 y m√°s a√±os por sexo y nivel de estudios‚Äù exportable a Excel ([Censos de Poblaci√≥n y Viviendas 2001 - INE](https://www.ine.es/censo/es/listatablas.jsp?group=5.Tablas%20comparativas%20de%20municipios%3E1.Personas%3E3.Estudios%20y%20estudiantes%3E0.Todos%20los%20municipios#:~:text=0.Todos%20los%20municipios.%20,la%20tabla%20en%20formato%20Excel)). El **Censo 2011** (basado parcialmente en muestra) tambi√©n ofrece datos municipales de nivel educativo, aunque con categor√≠as agregadas. Estas tablas est√°n disponibles en INEbase (formato PX/Excel) y via portal de datos abiertos (por ejemplo, Esri Espa√±a public√≥ un conjunto con el nivel de estudios del Censo 2011 por municipio) ([ArcGIS Hub](https://hub.arcgis.com/search?collection=Dataset&tags=censo%25202011#:~:text=Censo%202011%2C%20Demograficos%20,CSV%20%283468)).

- **Censo anual de poblaci√≥n ‚Äì Educaci√≥n (INE)** ‚Äì A partir de 2021 el INE ha comenzado a producir un **censo continuo** utilizando padrones y registros administrativos, con datos anuales por municipio. En 2024 difundi√≥ por primera vez la **estad√≠stica anual de poblaci√≥n por nivel de formaci√≥n** para **2021 y 2022** por municipio (excluyendo municipios <50 hab.) ([datos.gob.es](https://datos.gob.es/es/catalogo/ea0010587-poblacion-de-15-y-mas-anos-por-pais-de-nacimiento-espana-extranjero-y-nivel-de-estudios-agrupado-identificador-api-666661.rdf#:~:text=INEBase%20Population%20aged%2015%20and,openxmlformats)) ([datos.gob.es](https://datos.gob.es/es/catalogo/ea0010587-poblacion-de-15-y-mas-anos-por-pais-de-nacimiento-espana-extranjero-y-nivel-de-estudios-agrupado-identificador-api-666661.rdf#:~:text=INEBase%20Population%20aged%2015%20and,officedocument.spreadsheetml.sheet%20XLSX%20application%2Fjson%20JSON)). Estos datos ofrecen la poblaci√≥n de 15+ a√±os seg√∫n nivel educativo *agrupado* (p.ej. **educaci√≥n primaria o inferior**, **secundaria 1¬™ etapa**, **secundaria 2¬™ o post-secundaria**, **educaci√≥n superior**) en cada municipio, con frecuencia anual ([datos.gob.es](https://datos.gob.es/es/catalogo/ea0010587-poblacion-de-15-y-mas-anos-por-pais-de-nacimiento-espana-extranjero-y-nivel-de-estudios-agrupado-identificador-api-666661.rdf#:~:text=INEBase%20Population%20aged%2015%20and,officedocument.spreadsheetml.sheet%20XLSX%20application%2Fjson%20JSON)). Es previsible que se actualicen sucesivamente hasta 2024. *Nota:* En municipios mayores (>= 500 hab.) el INE permite un desglose m√°s detallado de niveles formativos, mientras que en los peque√±os se agrupan categor√≠as para salvaguardar el secreto estad√≠stico.

- **Estad√≠sticas auton√≥micas (ejemplo: IDESCAT)** ‚Äì Algunos institutos regionales ofrecen series anuales propias a nivel municipal aprovechando padrones y registros educativos. En Catalu√±a, el **Idescat** publica la *Estad√≠stica dels estudis de la poblaci√≥ (EEP)*, con la distribuci√≥n de la poblaci√≥n de 15+ a√±os por nivel de formaci√≥n alcanzado en cada municipio catal√°n. Esta serie anual (iniciada tras el censo 2011) se actualiza regularmente ([Idescat. Estad√≠stica dels estudis de la poblaci√≥. Catalunya](https://www.idescat.cat/pub/?id=eep#:~:text=Les%20dades%20s%27obtenen%20a%20partir,l%27aprofitament%20estad%C3%ADstic%20de%20registres%20administratius)). Idescat proporciona los datos en formatos abiertos (CSV enriquecido y JSON-Stat) para todos los municipios de Catalu√±a ([Idescat. Open Data. Educational Attainment Statistics](https://www.idescat.cat/dades/obertes/eep?lang=en#:~:text=CSV.%20Comma,Description)). Otras comunidades pueden tener iniciativas similares (p. ej., el Sistema de Informaci√≥n Multiterritorial de Andaluc√≠a ‚Äì SIMA ‚Äì difunde datos de censos 2001/2011 por municipio ([Sistema de Informaci√≥n Multiterritorial de Andaluc√≠a (SIMA)](https://www.juntadeandalucia.es/institutodeestadisticaycartografia/badea/informe/anual?CodOper=b3_151&idNode=23204#:~:text=Sistema%20de%20Informaci%C3%B3n%20Multiterritorial%20de,Censo)), y el instituto vasco **Eustat** publica indicadores municipales, aunque la disponibilidad de series hist√≥ricas anuales completas var√≠a).

- **Indicadores municipales abiertos** ‚Äì Adem√°s de las fuentes oficiales primarias, existen portales de datos abiertos que consolidan informaci√≥n municipal. El portal nacional [**datos.gob.es**](https://datos.gob.es) recopila conjuntos del INE, incluyendo tablas del censo por municipio. Por ejemplo, la tabla ‚ÄúPoblaci√≥n de 15 y m√°s a√±os por nivel de estudios (agrupado)‚Äù a nivel municipal est√° accesible con identificador API y descarga en CSV/Excel ([datos.gob.es](https://datos.gob.es/es/catalogo/ea0010587-poblacion-de-15-y-mas-anos-por-pais-de-nacimiento-espana-extranjero-y-nivel-de-estudios-agrupado-identificador-api-666661.rdf#:~:text=INEBase%20Population%20aged%2015%20and,officedocument.spreadsheetml.sheet%20XLSX%20application%2Fjson%20JSON)). Asimismo, plataformas como **ArcGIS Hub** (Esri) han publicado datasets listos para usar (CSV, shapefiles) con datos de nivel educativo por municipio del Censo 2011 ([ArcGIS Hub](https://hub.arcgis.com/search?collection=Dataset&tags=censo%25202011#:~:text=Censo%202011%2C%20Demograficos%20,CSV%20%283468)). Estas fuentes pueden facilitar la obtenci√≥n masiva sin tener que procesar manualmente los ficheros originales.

### üìã PLAN DE LIMPIEZA

| **Columna original**                       | **Transformaci√≥n**      | **Nueva columna(s)**                         | **Descripci√≥n**                                                                 |
|--------------------------------------------|--------------------------|-----------------------------------------------|----------------------------------------------------------------------------------|
| `Comunidades y Ciudades Aut√≥nomas`         | Filtrar                  | ‚Äî                                             | Eliminar filas donde el valor sea `"Total Nacional"`.                           |
| `Comunidades y Ciudades Aut√≥nomas`         | Separar                  | `ccaa_code`, `ccaa_name`                      | Separar c√≥digo y nombre. Ej: `01 Andaluc√≠a` ‚Üí `01`, `Andaluc√≠a`.                |
| `Total`                                    | Reemplazar ',' por '.'   | ‚Äî                                             | Normalizar formato decimal para convertir a float.                              |
| `Total`                                    | Convertir tipo           | `total`                                       | Convertir a tipo num√©rico (`float`). Filtrar filas con valores no convertibles. |
| `Total`                                    | Renombrar                | `total`                                       | Homogeneizar nombre.                                                            |
| `Sexo`, `Nivel de formaci√≥n alcanzado`     | Pasar a min√∫sculas       | `sexo`, `nivel_formacion`                     | Estandarizar nombres.                                                           |
| `Periodo`                                  | Pasar a min√∫sculas       | `periodo`                                     | Estandarizar nombres.                                                           |

---


````python
# Copia del DataFrame original
df_clean = df.copy()

# 1. Filtrar 'Total Nacional'
df_clean = df_clean[df_clean['Comunidades y Ciudades Aut√≥nomas'] != 'Total Nacional']

# 2. Separar c√≥digo y nombre de comunidad aut√≥noma
df_clean[['ccaa_code', 'ccaa_name']] = df_clean['Comunidades y Ciudades Aut√≥nomas'].str.extract(r'(\d+)\s+(.*)')

# 3. Reemplazar comas por puntos en 'Total' y convertir a float
df_clean['total'] = df_clean['Total'].str.replace(',', '.', regex=False)
df_clean = df_clean[df_clean['total'] != '..']  # Eliminar posibles valores no num√©ricos
df_clean['total'] = df_clean['total'].astype(float)

# 4. Renombrar y convertir nombres de columnas a min√∫sculas
df_clean.rename(columns={
    'Sexo': 'sexo',
    'Nivel de formaci√≥n alcanzado': 'nivel_formacion',
    'Periodo': 'periodo'
}, inplace=True)

# Eliminamos columna original ya procesada
df_clean.drop(columns=['Comunidades y Ciudades Aut√≥nomas', 'Total'], inplace=True)
df_clean
````

````python
# Eliminar filas donde 'nivel_formacion' es 'Total'
df_clean = df_clean[df_clean['nivel_formacion'].str.lower() != 'total']
````

````python
def print_unique_values_and_len(df):
    """Imprime los valores √∫nicos de cada columna en un DataFrame de Pandas y su longitud."""
    for column in df.columns:
        unique_values = df[column].unique()
        nan_count = df[column].isnull().sum()
        dtype = df[column].dtype

        print(f"Valores √∫nicos de la columna '{column}':")
        print(unique_values)
        print(f"Longitud de valores √∫nicos: {len(unique_values)}")
        print(f"N√∫mero de NaN: {nan_count}")
        print(f"Tipo de dato de la columna: {dtype}")
        print()

print_unique_values_and_len(df_clean)
````

````python
periodo_recuperado = df_clean.loc[df_clean.index, 'periodo']

# Separar en 'a√±o' y 'trimestre'
df_clean['a√±o'] = periodo_recuperado.str.extract(r'(\d{4})').astype(int)
df_clean['trimestre'] = periodo_recuperado.str.extract(r'(T\d)')
````

### ambos sexos

````python
df_ambos_sexos = df_clean[df_clean['sexo'].str.lower() == 'ambos sexos'].copy()
df_ambos_sexos
````

````python
# Agrupar por c√≥digo de comunidad, a√±o y nivel_formacion y calcular la media de 'total'
df_media_anual = (
    df_ambos_sexos
    .groupby(['ccaa_code', 'ccaa_name', 'a√±o', 'nivel_formacion'], as_index=False)
    .agg(media_total=('total', 'mean'))
)
df_media_anual
````

### df_media_anual

````python

# Filtra el DataFrame
df_filtered = df_media_anual[
    (df_media_anual['ccaa_code'] == "01") &
    (df_media_anual['nivel_formacion'] == 'Analfabetos')
]

# Muestra el DataFrame filtrado
df_filtered
````

````python
import matplotlib.pyplot as plt

# Agrupar por nivel de formaci√≥n y a√±o, promediando entre comunidades
df_comparacion = (
    df_media_anual
    .groupby(['nivel_formacion', 'a√±o'], as_index=False)
    .agg(media_total=('media_total', 'mean'))
)

# Crear gr√°fico comparativo
plt.figure(figsize=(12, 10))
for nivel in df_comparacion['nivel_formacion'].unique():
    datos = df_comparacion[df_comparacion['nivel_formacion'] == nivel]
    plt.plot(datos['a√±o'], datos['media_total'], marker='o', label=nivel)

plt.xlabel('A√±o')
plt.ylabel('Media total (%)')
plt.title('Evoluci√≥n del nivel de formaci√≥n por a√±o (media nacional)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
# Crear diccionario de asignaci√≥n de c√≥digos seg√∫n el orden propuesto
codigo_niveles = {
    'Analfabetos': 1,
    'Estudios primarios incompletos': 2,
    'Educaci√≥n primaria': 3,
    'Primera etapa de educaci√≥n secundaria y similar': 4,
    'Segunda etapa de educaci√≥n secundaria, con orientaci√≥n general': 5,
    'Segunda etapa de educaci√≥n secundaria con orientaci√≥n profesional (incluye educaci√≥n postsecundaria no superior)': 6,
    'Educaci√≥n superior': 7
}

# Asignar el c√≥digo correspondiente a cada nivel
df_media_anual['nivel_formacion_code'] = df_media_anual['nivel_formacion'].map(codigo_niveles)
df_media_anual
````

### df_formato_final

````python
# Crear tabla pivote con ccaa_code, a√±o y columnas 1 a 7 seg√∫n el c√≥digo de nivel_formacion
df_formato_final = df_media_anual.pivot_table(
    index=['ccaa_code', 'a√±o'],
    columns='nivel_formacion_code',
    values='media_total'
).reset_index()

# Renombrar columnas num√©ricas como strings
df_formato_final.columns.name = None
df_formato_final.columns = ['ccaa_code', 'a√±o'] + [str(col) for col in df_formato_final.columns[2:]]
df_formato_final
````

### guardamos

````python
# Crear un DataFrame para cada a√±o, eliminando la columna 'a√±o' en cada uno
dataframes_por_a√±o = {}
for a√±o in df_formato_final['a√±o'].unique():
    df_anual = df_formato_final[df_formato_final['a√±o'] == a√±o].drop(columns='a√±o')
    nombre = f"nivel_educativo_comunidades{a√±o}"
    dataframes_por_a√±o[nombre] = df_anual
    # Guardar como CSV
    df_anual.to_csv(f"/content/data final/{nombre}.csv", index=False)

# Mostrar los nombres de los dataframes creados
list(dataframes_por_a√±o.keys())
````

## üìú Informe de Procesamiento y Limpieza de Datos: Nivel Educativo por Comunidad Aut√≥noma
