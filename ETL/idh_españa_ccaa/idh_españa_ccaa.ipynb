```python
import pandas as pd

# URL de la tabla
url = "https://globaldatalab.org/shdi/table/shdi/ESP/?levels=1+4&interpolation=0&extrapolation=0"

# Intenta leer la tabla directamente desde la URL (puede fallar si la web no lo permite)
try:
    df = pd.read_html(url)[0]  # read_html devuelve una lista de DataFrames
    print("Tabla extraída exitosamente.")
    print(df.head())

except ValueError:
    print("Error: No se pudo leer la tabla directamente desde la URL.")
    print("Prueba a usar un web scraper más robusto si es necesario.")
```

```python
df = df.iloc[2:]
df
```

```python
df.rename(columns={df.columns[0]: 'comunidad_name'}, inplace=True)
df
```

### datos comunidades

```python
import pandas as pd
import warnings

# Suprime los warnings
warnings.filterwarnings('ignore')

# URL y lectura de la primera tabla
url = 'https://www.ine.es/daco/daco42/codmun/cod_ccaa_provincia.htm'
df_temp = pd.read_html(url, encoding='ISO-8859-1')[0]
df_temp.columns = ['CODAUTO', 'Comunidad Autónoma', 'CPRO', 'Provincia']

# Filtra para eliminar "Ciudades Autónomas" y convierte CODAUTO a numérico
df_temp = df_temp[df_temp['Comunidad Autónoma'] != 'Ciudades Autónomas']
df_temp['CODAUTO'] = pd.to_numeric(df_temp['CODAUTO'], errors='coerce')
df_temp = df_temp.dropna(subset=['CODAUTO'])

# Guarda el DataFrame en CSV y muestra las primeras filas
df_temp.to_csv('codigos_ccaa_provincias.csv', index=False, encoding='utf-8')

# Crea y muestra el DataFrame de equivalencias de comunidades
equivalencias_comunidades = df_temp[['CODAUTO', 'Comunidad Autónoma']].drop_duplicates('CODAUTO')
print(equivalencias_comunidades)
```

```python
df_map = df.copy()

df_map['comunidad_name'] = df_map['comunidad_name'].str.replace('ciudad Autonoma de', '', case=False)
df_map['comunidad_name'] = df_map['comunidad_name'].str.replace('Region', '', case=False)


import pandas as pd
from fuzzywuzzy import fuzz


def fuzzy_mapper(x, mapping_series):
    """Maps a value to the closest match in a given series using fuzzy matching."""
    best_match = None
    best_score = 40

    for key, value in mapping_series.items():
      score = fuzz.ratio(x, key)
      if score > best_score:
        best_score = score
        best_match = value
    return best_match


df_map['CODAUTO'] = df_map['comunidad_name'].apply(lambda x: fuzzy_mapper(x, equivalencias_comunidades.set_index('Comunidad Autónoma')['CODAUTO']))

df_map = df_map.drop(columns=['comunidad_name'])

# Merge the dataframes
df_merged = pd.merge(df_map, equivalencias_comunidades, on='CODAUTO', how='left')
# Rename the 'Comunidad Autónoma' column to 'comunidad_name' and drop the original one.
df_merged = df_merged.rename(columns={'Comunidad Autónoma': 'comunidad_name'})
df_merged
```

```python
df_merged.to_csv('idh_españa_completo.csv', encoding='utf-8-sig', index=False)
files.download('idh_españa_completo.csv')
```

```python
import matplotlib.pyplot as plt

# Renombrar el dataframe como df_merged (ya lo es)
# Eliminar columna 'comunidad_name.1' si aún está presente
if 'comunidad_name.1' in df_merged.columns:
    df_merged = df_merged.drop(columns=['comunidad_name.1'])

# Transformar el DataFrame a formato largo para graficar la serie histórica
df_long = df_merged.melt(id_vars=['comunidad_name'],
                         var_name='Año',
                         value_name='Valor')

# Convertir 'Año' a numérico
df_long['Año'] = pd.to_numeric(df_long['Año'], errors='coerce')

# Graficar
plt.figure(figsize=(14, 8))
for comunidad in df_long['comunidad_name'].unique():
    datos = df_long[df_long['comunidad_name'] == comunidad]
    plt.plot(datos['Año'], datos['Valor'], label=comunidad)

plt.title(' IDH - Serie histórica por comunidad autónoma')
plt.xlabel('Año')
plt.ylabel('Valor')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

## 🧾 Informe de Obtención y Procesamiento del Índice de Desarrollo Humano (IDH) en España

### 🧭 Introducción

Este informe describe el proceso seguido para la obtención del Índice de Desarrollo Humano (IDH) en España, con el objetivo de adaptarlo al formato de simulación utilizado por *PolicySpace2*, originalmente diseñado para datos a nivel municipal en Brasil. El propósito principal es sentar las bases para una posible replicación del análisis de políticas públicas en el contexto español.

### 🔎 Proceso de Obtención

| Paso                           | Descripción                                                                                             |
|--------------------------------|---------------------------------------------------------------------------------------------------------|
| Fuente principal               | Se utilizaron los datos del **Global Data Lab**: https://globaldatalab.org/shdi/table/shdi/ESP         |
| Nivel de desagregación         | Los datos disponibles están a **nivel de comunidad autónoma**                                           |
| Formato original               | Los datos se extrajeron directamente en formato tabla utilizando `pandas.read_html()`                   |
| Normalización de nombres      | Se aplicaron transformaciones como eliminación de "Región", "Ciudad Autónoma de", etc.                 |
| Correspondencia codificada     | Se empleó `fuzzy matching` para asociar nombres con sus códigos oficiales del INE                       |
| Unificación con datos del INE | Se descargó y preprocesó la tabla de correspondencia CCAA–provincia desde el sitio del INE              |
| Guardado de resultados         | Se generó y descargó el CSV final como `idh_españa_completo.csv`                                        |

### 🧼 Observaciones Importantes

- El nivel de desagregación alcanzado actualmente es **comunitario**, mientras que en el modelo original brasileño se trabaja con **datos municipales** (`cod_mun`).
- Esto representa una limitación importante para simular políticas locales con el nivel de granularidad esperado por *PolicySpace2*.
- Por tanto, **una recomendación clave es desarrollar una metodología que permita estimar el IDH a nivel municipal**. Para ello se podrían utilizar:
  - Modelos de interpolación o estimación a partir de indicadores municipales disponibles (renta, educación, esperanza de vida).
  - Redistribución proporcional basada en población, nivel educativo y renta media por municipio.
  - Uso de proxies alternativos o datos del padrón, renta bruta municipal o similares.
