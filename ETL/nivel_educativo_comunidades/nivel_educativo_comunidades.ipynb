## Fuentes de datos por municipio (2000–2024)

Para obtener el nivel de **cualificación educativa de la población por municipio** en España, las fuentes más confiables son los **censos de población** y estadísticas oficiales derivadas, complementadas por datos autonómicos y portales abiertos:

- **Censos de Población (INE)** – Los censos nacionales proporcionan puntos de datos clave (aprox. decenales) con **detalle municipal**. En el **Censo 2001** y **Censo 2011** se publicó la población por nivel de estudios alcanzado en cada municipio (habitualmente para población de 16 o 15 años y más, desglosada por categorías educativas) ([Censos de Población y Viviendas 2001 - INE](https://www.ine.es/censo/es/listatablas.jsp?group=5.Tablas%20comparativas%20de%20municipios%3E1.Personas%3E3.Estudios%20y%20estudiantes%3E0.Todos%20los%20municipios#:~:text=0.Todos%20los%20municipios.%20,la%20tabla%20en%20formato%20Excel)). El INE difundió tablas descargables por municipio; p. ej., para 2001 existe una tabla “Todos los municipios – población de 16 y más años por sexo y nivel de estudios” exportable a Excel ([Censos de Población y Viviendas 2001 - INE](https://www.ine.es/censo/es/listatablas.jsp?group=5.Tablas%20comparativas%20de%20municipios%3E1.Personas%3E3.Estudios%20y%20estudiantes%3E0.Todos%20los%20municipios#:~:text=0.Todos%20los%20municipios.%20,la%20tabla%20en%20formato%20Excel)). El **Censo 2011** (basado parcialmente en muestra) también ofrece datos municipales de nivel educativo, aunque con categorías agregadas. Estas tablas están disponibles en INEbase (formato PX/Excel) y via portal de datos abiertos (por ejemplo, Esri España publicó un conjunto con el nivel de estudios del Censo 2011 por municipio) ([ArcGIS Hub](https://hub.arcgis.com/search?collection=Dataset&tags=censo%25202011#:~:text=Censo%202011%2C%20Demograficos%20,CSV%20%283468)).

- **Censo anual de población – Educación (INE)** – A partir de 2021 el INE ha comenzado a producir un **censo continuo** utilizando padrones y registros administrativos, con datos anuales por municipio. En 2024 difundió por primera vez la **estadística anual de población por nivel de formación** para **2021 y 2022** por municipio (excluyendo municipios <50 hab.) ([datos.gob.es](https://datos.gob.es/es/catalogo/ea0010587-poblacion-de-15-y-mas-anos-por-pais-de-nacimiento-espana-extranjero-y-nivel-de-estudios-agrupado-identificador-api-666661.rdf#:~:text=INEBase%20Population%20aged%2015%20and,openxmlformats)) ([datos.gob.es](https://datos.gob.es/es/catalogo/ea0010587-poblacion-de-15-y-mas-anos-por-pais-de-nacimiento-espana-extranjero-y-nivel-de-estudios-agrupado-identificador-api-666661.rdf#:~:text=INEBase%20Population%20aged%2015%20and,officedocument.spreadsheetml.sheet%20XLSX%20application%2Fjson%20JSON)). Estos datos ofrecen la población de 15+ años según nivel educativo *agrupado* (p.ej. **educación primaria o inferior**, **secundaria 1ª etapa**, **secundaria 2ª o post-secundaria**, **educación superior**) en cada municipio, con frecuencia anual ([datos.gob.es](https://datos.gob.es/es/catalogo/ea0010587-poblacion-de-15-y-mas-anos-por-pais-de-nacimiento-espana-extranjero-y-nivel-de-estudios-agrupado-identificador-api-666661.rdf#:~:text=INEBase%20Population%20aged%2015%20and,officedocument.spreadsheetml.sheet%20XLSX%20application%2Fjson%20JSON)). Es previsible que se actualicen sucesivamente hasta 2024. *Nota:* En municipios mayores (>= 500 hab.) el INE permite un desglose más detallado de niveles formativos, mientras que en los pequeños se agrupan categorías para salvaguardar el secreto estadístico.

- **Estadísticas autonómicas (ejemplo: IDESCAT)** – Algunos institutos regionales ofrecen series anuales propias a nivel municipal aprovechando padrones y registros educativos. En Cataluña, el **Idescat** publica la *Estadística dels estudis de la població (EEP)*, con la distribución de la población de 15+ años por nivel de formación alcanzado en cada municipio catalán. Esta serie anual (iniciada tras el censo 2011) se actualiza regularmente ([Idescat. Estadística dels estudis de la població. Catalunya](https://www.idescat.cat/pub/?id=eep#:~:text=Les%20dades%20s%27obtenen%20a%20partir,l%27aprofitament%20estad%C3%ADstic%20de%20registres%20administratius)). Idescat proporciona los datos en formatos abiertos (CSV enriquecido y JSON-Stat) para todos los municipios de Cataluña ([Idescat. Open Data. Educational Attainment Statistics](https://www.idescat.cat/dades/obertes/eep?lang=en#:~:text=CSV.%20Comma,Description)). Otras comunidades pueden tener iniciativas similares (p. ej., el Sistema de Información Multiterritorial de Andalucía – SIMA – difunde datos de censos 2001/2011 por municipio ([Sistema de Información Multiterritorial de Andalucía (SIMA)](https://www.juntadeandalucia.es/institutodeestadisticaycartografia/badea/informe/anual?CodOper=b3_151&idNode=23204#:~:text=Sistema%20de%20Informaci%C3%B3n%20Multiterritorial%20de,Censo)), y el instituto vasco **Eustat** publica indicadores municipales, aunque la disponibilidad de series históricas anuales completas varía).

- **Indicadores municipales abiertos** – Además de las fuentes oficiales primarias, existen portales de datos abiertos que consolidan información municipal. El portal nacional [**datos.gob.es**](https://datos.gob.es) recopila conjuntos del INE, incluyendo tablas del censo por municipio. Por ejemplo, la tabla “Población de 15 y más años por nivel de estudios (agrupado)” a nivel municipal está accesible con identificador API y descarga en CSV/Excel ([datos.gob.es](https://datos.gob.es/es/catalogo/ea0010587-poblacion-de-15-y-mas-anos-por-pais-de-nacimiento-espana-extranjero-y-nivel-de-estudios-agrupado-identificador-api-666661.rdf#:~:text=INEBase%20Population%20aged%2015%20and,officedocument.spreadsheetml.sheet%20XLSX%20application%2Fjson%20JSON)). Asimismo, plataformas como **ArcGIS Hub** (Esri) han publicado datasets listos para usar (CSV, shapefiles) con datos de nivel educativo por municipio del Censo 2011 ([ArcGIS Hub](https://hub.arcgis.com/search?collection=Dataset&tags=censo%25202011#:~:text=Censo%202011%2C%20Demograficos%20,CSV%20%283468)). Estas fuentes pueden facilitar la obtención masiva sin tener que procesar manualmente los ficheros originales.

### 📋 PLAN DE LIMPIEZA

| **Columna original**                       | **Transformación**      | **Nueva columna(s)**                         | **Descripción**                                                                 |
|--------------------------------------------|--------------------------|-----------------------------------------------|----------------------------------------------------------------------------------|
| `Comunidades y Ciudades Autónomas`         | Filtrar                  | —                                             | Eliminar filas donde el valor sea `"Total Nacional"`.                           |
| `Comunidades y Ciudades Autónomas`         | Separar                  | `ccaa_code`, `ccaa_name`                      | Separar código y nombre. Ej: `01 Andalucía` → `01`, `Andalucía`.                |
| `Total`                                    | Reemplazar ',' por '.'   | —                                             | Normalizar formato decimal para convertir a float.                              |
| `Total`                                    | Convertir tipo           | `total`                                       | Convertir a tipo numérico (`float`). Filtrar filas con valores no convertibles. |
| `Total`                                    | Renombrar                | `total`                                       | Homogeneizar nombre.                                                            |
| `Sexo`, `Nivel de formación alcanzado`     | Pasar a minúsculas       | `sexo`, `nivel_formacion`                     | Estandarizar nombres.                                                           |
| `Periodo`                                  | Pasar a minúsculas       | `periodo`                                     | Estandarizar nombres.                                                           |

````python
# Copia del DataFrame original
df_clean = df.copy()

# 1. Filtrar 'Total Nacional'
df_clean = df_clean[df_clean['Comunidades y Ciudades Autónomas'] != 'Total Nacional']

# 2. Separar código y nombre de comunidad autónoma
df_clean[['ccaa_code', 'ccaa_name']] = df_clean['Comunidades y Ciudades Autónomas'].str.extract(r'(\d+)\s+(.*)')

# 3. Reemplazar comas por puntos en 'Total' y convertir a float
df_clean['total'] = df_clean['Total'].str.replace(',', '.', regex=False)
df_clean = df_clean[df_clean['total'] != '..']  # Eliminar posibles valores no numéricos
df_clean['total'] = df_clean['total'].astype(float)

# 4. Renombrar y convertir nombres de columnas a minúsculas
df_clean.rename(columns={
    'Sexo': 'sexo',
    'Nivel de formación alcanzado': 'nivel_formacion',
    'Periodo': 'periodo'
}, inplace=True)

# Eliminamos columna original ya procesada
df_clean.drop(columns=['Comunidades y Ciudades Autónomas', 'Total'], inplace=True)
df_clean
````

````python
# Eliminar filas donde 'nivel_formacion' es 'Total'
df_clean = df_clean[df_clean['nivel_formacion'].str.lower() != 'total']
````

````python
def print_unique_values_and_len(df):
    """Imprime los valores únicos de cada columna en un DataFrame de Pandas y su longitud."""
    for column in df.columns:
        unique_values = df[column].unique()
        nan_count = df[column].isnull().sum()
        dtype = df[column].dtype

        print(f"Valores únicos de la columna '{column}':")
        print(unique_values)
        print(f"Longitud de valores únicos: {len(unique_values)}")
        print(f"Número de NaN: {nan_count}")
        print(f"Tipo de dato de la columna: {dtype}")
        print()

print_unique_values_and_len(df_clean)
````

````python
periodo_recuperado = df_clean.loc[df_clean.index, 'periodo']

# Separar en 'año' y 'trimestre'
df_clean['año'] = periodo_recuperado.str.extract(r'(\d{4})').astype(int)
df_clean['trimestre'] = periodo_recuperado.str.extract(r'(T\d)')
````

### ambos sexos

````python
df_ambos_sexos = df_clean[df_clean['sexo'].str.lower() == 'ambos sexos'].copy()
df_ambos_sexos
````

````python
# Agrupar por código de comunidad, año y nivel_formacion y calcular la media de 'total'
df_media_anual = (
    df_ambos_sexos
    .groupby(['ccaa_code', 'ccaa_name', 'año', 'nivel_formacion'], as_index=False)
    .agg(media_total=('total', 'mean'))
)
df_media_anual
````

### df_media_anual

````python

# Filtra el DataFrame
df_filtered = df_media_anual[
    (df_media_anual['ccaa_code'] == "01") &
    (df_media_anual['nivel_formacion'] == 'Analfabetos')
]

# Muestra el DataFrame filtrado
df_filtered
````

````python
import matplotlib.pyplot as plt

# Agrupar por nivel de formación y año, promediando entre comunidades
df_comparacion = (
    df_media_anual
    .groupby(['nivel_formacion', 'año'], as_index=False)
    .agg(media_total=('media_total', 'mean'))
)

# Crear gráfico comparativo
plt.figure(figsize=(12, 10))
for nivel in df_comparacion['nivel_formacion'].unique():
    datos = df_comparacion[df_comparacion['nivel_formacion'] == nivel]
    plt.plot(datos['año'], datos['media_total'], marker='o', label=nivel)

plt.xlabel('Año')
plt.ylabel('Media total (%)')
plt.title('Evolución del nivel de formación por año (media nacional)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
````

````python
# Crear diccionario de asignación de códigos según el orden propuesto
codigo_niveles = {
    'Analfabetos': 1,
    'Estudios primarios incompletos': 2,
    'Educación primaria': 3,
    'Primera etapa de educación secundaria y similar': 4,
    'Segunda etapa de educación secundaria, con orientación general': 5,
    'Segunda etapa de educación secundaria con orientación profesional (incluye educación postsecundaria no superior)': 6,
    'Educación superior': 7
}

# Asignar el código correspondiente a cada nivel
df_media_anual['nivel_formacion_code'] = df_media_anual['nivel_formacion'].map(codigo_niveles)
df_media_anual
````

### df_formato_final

````python
# Crear tabla pivote con ccaa_code, año y columnas 1 a 7 según el código de nivel_formacion
df_formato_final = df_media_anual.pivot_table(
    index=['ccaa_code', 'año'],
    columns='nivel_formacion_code',
    values='media_total'
).reset_index()

# Renombrar columnas numéricas como strings
df_formato_final.columns.name = None
df_formato_final.columns = ['ccaa_code', 'año'] + [str(col) for col in df_formato_final.columns[2:]]
df_formato_final
````

### guardamos

````python
# Crear un DataFrame para cada año, eliminando la columna 'año' en cada uno
dataframes_por_año = {}
for año in df_formato_final['año'].unique():
    df_anual = df_formato_final[df_formato_final['año'] == año].drop(columns='año')
    nombre = f"nivel_educativo_comunidades{año}"
    dataframes_por_año[nombre] = df_anual
    # Guardar como CSV
    df_anual.to_csv(f"/content/data final/{nombre}.csv", index=False)

# Mostrar los nombres de los dataframes creados
list(dataframes_por_año.keys())
````

## 📜 Informe de Procesamiento y Limpieza de Datos: Nivel Educativo por Comunidad Autónoma

### 🔖 Introducción
Este informe documenta el proceso de limpieza, transformación y generación de datos relativos al nivel educativo de la población por comunidad autónoma en España. A partir de datos originales proporcionados por el INE, se ha estructurado una base de datos homogénea y comparativa para el análisis anual entre los años 2014 y 2024.

A diferencia del modelo `PolicySpace2`, que trabaja con 15 categorías de cualificación y lo hace sobre áreas ponderadas de Brasil, nuestro enfoque utiliza **solo 7 niveles educativos** estandarizados y se centra en **comunidades autónomas españolas**, permitiendo una adaptación directa al contexto nacional.

Esta diferencia de resolución territorial (comunidad vs área ponderada) tiene implicaciones importantes para su uso en simulaciones:
- ⚠️ **Menor granularidad espacial:** No permite analizar desigualdades intracomunitarias o dinámicas locales urbanas/rurales, como hace `PolicySpace2` con zonas dentro de las regiones metropolitanas.
- ✅ **Mayor comparabilidad y aplicabilidad nacional:** Nuestros datos sí permiten simular políticas educativas o de redistribución a nivel nacional o autonómico con coherencia estadística.
- 🧠 **Modelo compatible con niveles agregados:** El modelo puede aprovecharse para estudiar escenarios de evolución educativa, migración por nivel formativo, o cambios estructurales del mercado laboral a nivel macro.
